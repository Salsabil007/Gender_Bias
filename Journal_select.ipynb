```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_csv("Altmetric_Journal.csv")
print(data.dtypes)
print(len(data))
#print(data.head(5))
```

```python
data = data.sort_values(by=['News mentions'], ascending = False)
print(data.head(5))
```

```python
df2 = pd.read_excel("sm_journal_classification.xlsx", 'Journal_Classification')
print(len(df2))
```

```python
print(df2.dtypes)
```

```python
domain = []
field = []
subfield = []

#data['Journal title'].apply(lambda x: x.lower())
data['Journal title'] = data['Journal title'].str.lower()
df2['Source_title'] = df2['Source_title'].str.lower()
#df2['Source_title'].apply(lambda x: x.lower())
#print(df2.head(5))

for ind in data.index:
    j = data['Journal title'][ind]
    f = df2[df2['Source_title'] == j]
    if len(f) > 1:
        #print("more than one ", j)
        '''domain.append("multiple")
        field.append("multiple")
        subfield.append("multiple")'''
        if pd.isna(data['ISSNs'][ind]):
            domain.append(f.iloc[0]['Domain_English'])
            field.append(f.iloc[0]['Field_English'])
            subfield.append(f.iloc[0]['SubField_English'])
        else:
            lst = [x.strip() for x in data['ISSNs'][ind].split(',')]
            newlst = [i.replace('-','') for i in lst]
            flag = 0
            for i in newlst:
                ff = f[f['issn1'] == i]
                if len(ff) >= 1:
                    #print("yes1 ",i)
                    domain.append(ff.iloc[0]['Domain_English'])
                    field.append(ff.iloc[0]['Field_English'])
                    subfield.append(ff.iloc[0]['SubField_English'])
                    flag = 1
                    break
                ff = f[f['issn2'] == i]
                if len(ff) >= 1:
                    #print("yes2 ",i)
                    domain.append(ff.iloc[0]['Domain_English'])
                    field.append(ff.iloc[0]['Field_English'])
                    subfield.append(ff.iloc[0]['SubField_English'])
                    flag = 1
                    break
                ff = f[f['issn3'] == i]
                if len(ff) >= 1:
                    #print("yes3 ",i)
                    domain.append(ff.iloc[0]['Domain_English'])
                    field.append(ff.iloc[0]['Field_English'])
                    subfield.append(ff.iloc[0]['SubField_English'])
                    flag = 1
                    break
            if flag == 0:
                domain.append("multiple")
                field.append("multiple")
                subfield.append("multiple")
                    
            
    elif len(f) == 1:
        domain.append(f.iloc[0]['Domain_English'])
        field.append(f.iloc[0]['Field_English'])
        subfield.append(f.iloc[0]['SubField_English'])
    else:
        
        if pd.isna(data['ISSNs'][ind]):
            domain.append("not found")
            field.append("not found")
            subfield.append("not found")
        else:
            lst = [x.strip() for x in data['ISSNs'][ind].split(',')]
            newlst = [i.replace('-','') for i in lst]
            flag = 0
            for i in newlst:
                ff = df2[df2['issn1'] == i]
                if len(ff) >= 1:
                    #print("yes1 ",i)
                    domain.append(ff.iloc[0]['Domain_English'])
                    field.append(ff.iloc[0]['Field_English'])
                    subfield.append(ff.iloc[0]['SubField_English'])
                    flag = 1
                    break
                ff = df2[df2['issn2'] == i]
                if len(ff) >= 1:
                    #print("yes2 ",i)
                    domain.append(ff.iloc[0]['Domain_English'])
                    field.append(ff.iloc[0]['Field_English'])
                    subfield.append(ff.iloc[0]['SubField_English'])
                    flag = 1
                    break
                ff = df2[df2['issn3'] == i]
                if len(ff) >= 1:
                    #print("yes3 ",i)
                    domain.append(ff.iloc[0]['Domain_English'])
                    field.append(ff.iloc[0]['Field_English'])
                    subfield.append(ff.iloc[0]['SubField_English'])
                    flag = 1
                    break
            if flag == 0:
                domain.append("not found")
                field.append("not found")
                subfield.append("not found")
data['domain'] = domain
data['field'] = field
data['subfield'] = subfield

final_data = data[data['domain'] != 'not found']
final_data = final_data[final_data['domain'] != 'multiple']
print(len(final_data))
```

```python
##extracting 25% journals for each field
final_data = pd.read_csv("journal_merged_with_classification.csv")
print(len(final_data))
final_data = final_data.dropna(subset=['ISSNs'])
final_data = final_data.dropna(subset=['Journal title'])
print(len(final_data))
unique_field = final_data['field'].unique()
data = pd.DataFrame()
for i in unique_field:
    temp = final_data[final_data['field'] == i]
    temp = temp.sort_values(by = ['News mentions'], ascending = False)
    l = len(temp)
    l = math.ceil(l * 0.25)
    tt = temp.head(l)
    #data = data.append(tt, ignore_index = True)
    data = pd.concat([data, tt], ignore_index=True)
```
